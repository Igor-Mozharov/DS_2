##### Отримайте список регіонів, де рівень народжуваності у 2014 році був вищим за середній по Україні

##### Побудуйте стовпчикову діаграму народжуваності по регіонах у 2014 році

```python
population[0]['2014'][:-1].plot(kind='bar')
plt.show()
```

# Візуалізація:

```python
population[0]['1950'][:-1].plot(kind='pie', fontsize='x-small')
plt.title('Population 1950', fontsize='xx-large', color='g', loc='right')
plt.show()
```

```python
population[0][['2008', '2012']].iloc[0:3].plot(fontsize='x-small')
plt.grid(True)
plt.xlabel('Years', color='b', fontsize='medium')
plt.ylabel('Population', color='b', fontsize='medium')
plt.legend(title='Years', loc='best')
plt.title('3 REGION POPULATION IN 2008, 2012 Y.', fontweight='heavy', color='purple')
plt.show()
```

```python
plt.stem(population[0]['1950'][:-1], linefmt=':', label='Population')
plt.ylabel('Population', color='purple', fontsize='x-large')
plt.xlabel('Numbers of regions', color='purple', fontsize='x-large')
plt.title('Difference', color='black', fontweight='heavy', fontsize='xx-large')
plt.legend()
plt.show()
```
