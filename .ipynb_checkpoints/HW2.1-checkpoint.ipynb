# Частина перша: Знайомство з Pandas

```python
population = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96', match='Коефіцієнт народжуваності у регіонах України у 1950—2014 роках', index_col=0)
```

##### Вивести перші рядки таблиці за допомогою методу head

```python
population_rep = population[0].replace(['-'], np.nan)
population_rep
```

##### Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
population[0].dtypes
```

##### Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

##### Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

##### Видаліть з таблиці дані по всій країні, останній рядок таблиці

##### Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
population[0].fillna(population[0].mean())
```

##### Отримайте список регіонів, де рівень народжуваності у 2014 році був вищим за середній по Україні

```python
population[0]['2014'].iloc[:-1][population[0]['2014'].iloc[:-1] > population[0]['2014'].iloc[-1].mean()]
```

##### У якому регіоні була найвища народжуваність у 2014 році?

```python
population[0]['2014'].idxmax()
```

##### Побудуйте стовпчикову діаграму народжуваності по регіонах у 2014 році
